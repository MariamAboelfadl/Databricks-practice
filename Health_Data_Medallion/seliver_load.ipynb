```python
%sql
create or replace view my_catalog.default.vw_silver_health as
select 
case when Country='USA' then 'United States' 
when Country='UK' then 'United Kingdom' else Country
 end as country,

`Disease Name`,`Disease Category`,`Age Group`,`treatment type`,
cast(`Recovery Rate` as double) as `Recovery Rate`,Gender,
cast(year as int ) as year,
cast(`Incidence Rate` as double) as `Incidence Rate`,
cast(`Mortality Rate` as double) as `Mortality Rate`,
cast(`Prevalence Rate` as double) as `prevalence_rate`,
cast(`Population Affected` as double) as `Population Affected`,
cast(`Healthcare Access` as double) as `Healthcare Access`,
cast(`Doctors per 1000` as double) as `Doctors per 1000`,
cast(`Hospital Beds per 1000` as double) as `Hospital Beds per 1000`,
cast(`Average Treatment Cost USD` as double) as `Average Treatment Cost USD` ,
cast(`improvement in 5 years`as double) as `improvement_in_5_years`,
`Availability of Vaccines Treatment`,
cast(`Mortality Rate` as double) / cast(`Incidence Rate` as double) as mortality_to_incidence_ratio,
cast(`Recovery Rate` as double) / cast(`Incidence Rate` as double) as recovery_to_incidence_ratio,

ingestion_time, source_file
from my_catalog.default.bronze_heatlh_tb where country is not null and year is not null and `Disease Name` is not null 


```

```python
%sql
drop  table if exists my_catalog.default.silver_health_tb ;

CREATE TABLE my_catalog.default.silver_health_tb
TBLPROPERTIES ('delta.columnMapping.mode' = 'name')
AS SELECT * FROM my_catalog.default.vw_silver_health WHERE 1=0;
```

```python

%sql
MERGE INTO my_catalog.default.silver_health_tb AS tgt
USING my_catalog.default.vw_silver_health AS src
ON tgt.country = src.country
  AND tgt.year = src.year
  AND tgt.`Disease Name` = src.`Disease Name`
WHEN MATCHED THEN
  UPDATE SET *
WHEN NOT MATCHED THEN
  INSERT *

```
