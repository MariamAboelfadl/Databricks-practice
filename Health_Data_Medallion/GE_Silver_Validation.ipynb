```python
import great_expectations as ge


df = spark.table("my_catalog.default.silver_health_tb").toPandas()


ge_df = ge.from_pandas(df)

#check nulls
ge_df.expect_column_values_to_not_be_null("country")
ge_df.expect_column_values_to_not_be_null("year")
ge_df.expect_column_values_to_not_be_null("Disease Name")

#check duplications
#ge_df.expect_compound_columns_to_be_unique(["country", "year", "Disease Name"])

#check valid and max/min values
ge_df.expect_column_values_to_be_between("year", 2000, 2025)
ge_df.expect_column_values_to_be_between("Incidence Rate", 0, 100)
ge_df.expect_column_values_to_be_between("Mortality Rate", 0, 100)
ge_df.expect_column_values_to_be_between("prevalence_rate", 0, 100)

#check important selected columns
ge_df.expect_table_columns_to_match_ordered_list([
    "country","Disease Name","Disease Category","Age Group",
    "treatment type","Recovery Rate","Gender","year",
    "Incidence Rate","Mortality Rate","prevalence_rate",
    "Population Affected","Healthcare Access","Doctors per 1000",
    "Hospital Beds per 1000","Average Treatment Cost USD",
    "improvement_in_5_years","Availability of Vaccines Treatment",
    "mortality_to_incidence_ratio","recovery_to_incidence_ratio",
    "ingestion_time","source_file"
])

# Validate
results = ge_df.validate()
display(results)
```
